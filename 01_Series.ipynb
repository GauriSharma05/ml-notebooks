PANDAS SERIES

Creating series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {2: 10, 4: 20, 6: 30}
```

```python
pd.Series(list)
#here pandas give default index values. the right column is called data point and the left column is the index points.
```

```python
pd.Series(list,labels)
#here we are giving our own index values. that is we customized the index values. the right column is called data point and the left column is the index points.
```

```python
pd.Series(arr)
#in same way we can create series from numpy array as we did from list.
```

```python
pd.Series(arr,labels)
#we can customize the index values for numpy array as well.
```

```python
pd.Series(d)
#when we create series from dictionary, the keys become the index values and the values become the data points. we need not mention them seprately
```
