```python
import numpy as np 
import pandas as pd
```

Creating a dataframe

```python
data ={
    "Name": ['A', 'B', 'C', 'D'],
    "Age": ['10', '20', '30', '40'],
    "City": ['jal', 'una', 'kullu', 'jaipur'],
    "Salary": ['25', '35', '45', '55']
}
```

```python
df =pd.DataFrame(data)
df
#in dictionary the columns are the keys and the values are the list of values for each column. The DataFrame is created by passing the dictionary to the pd.DataFrame() function. 
```

```python
list = [
    ['A', '10', 'jal', '25'],
    ['B', '20', 'una', '35'],
    ['C', '30', 'kullu', '45'],
    ['D', '40', 'jaipur', '55']
]
labels = ['Name', 'Age', 'City', 'Salary']
```

```python
df2 = pd.DataFrame(list)
df2
#in list column names are not provided, so pandas assigns default column names as 0, 1, 2, 3. 
```

```python
df3 = pd.DataFrame(list, columns=labels)
df3
#in list we can customize the column names by passing a list of column names to the columns parameter of the pd.DataFrame() function.
```
