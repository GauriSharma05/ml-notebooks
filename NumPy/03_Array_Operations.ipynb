```python
a1*a2
```

```python
a1/a2
```

```python
a1//a2
```

Broadcasting

```python
l = [10,20,30,40]
a = np.array(l)
```

```python
a + 10
```

```python
for i in l:
    print(i+10)
```

```python
a3 = np.arange(1,26).reshape(5,5)
a3
```

```python
a3+10
```

```python
a4 = np.arange(1,10).reshape(3,3)
a4
```
