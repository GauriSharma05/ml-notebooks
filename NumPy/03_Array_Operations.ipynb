ADVANCE ARRAY MANIPULATION

Stacking array

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a, b))
```

```python
np.vstack([a, b])
```

```python
np.hstack([a, b])
```

```python
np.hstack((a, b))
```

```python
np.column_stack((a, b))
```

```python
np.column_stack([a, b])
```

```python
np.row_stack((a,b))
```

```python
np.stack((a, b), axis=0)
#axis=0 means it will stack the arrays vertically. same as np.vstack((a,b))
```

```python
print(np.stack((a,b), axis=1))
#axis=1 means it will stack the arrays horizontally. same as np.column_stack((a,b))
```

```python
ab = np.array([[1,2],
              [3,4]])

ba = np.array([[5,6],
              [7,8]])

print(np.dstack((ab,ba)))
#it will stack the arrays depth wise. It will create a new dimension. The first array will be in the first layer and the second array will be in the second layer.
```
