```python
import numpy as np
```

Creating arrays

```python
arr = np.array([1, 2, 3, 4, 5])
arr
```

```python
a = [1, 2, 3, 4, 5, "hello"]
a
```

```python
np.array(a)
#in arrays all elements must be of the same data type.  If we try to create an array with mixed data types, NumPy will upcast the elements to a common data type that can accommodate all the values. In this case, since "hello" is a string, all the other elements will be converted to strings as well.
```

```python
b = [1,2,3.5]
b
#In this case, since 3.5 is a float, all the other elements will be converted to floats as well.
```

```python
np.array(b)
```

```python
d = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
d
#lists can be nested to create multi-dimensional arrays. In this case, we have a 2D list with three rows and three columns. but lists are vectors
```

```python
np.array(d)
# arrays can also be multi-dimensional. In this case, we have a 2D array with three rows and three columns. Unlike lists, arrays are homogeneous, meaning that all elements must be of the same data type.
```

```python
np.arange(5,0,-1)
# arange() function is used to create an array with a range of values. It takes three arguments: start, stop, and step. The start argument specifies the starting value of the range, the stop argument specifies the ending value of the range (exclusive), and the step argument specifies the increment between each value in the range. If the step argument is negative, the range will be created in reverse order.
```

```python
np.linspace(0, 0.4, 5)
```

```python
arr = np.arange(1,11,2)
arr
```

```python
arr = np.zeros(5)
arr
#zeros() function is used to create an array filled with zeros. It takes a single argument, which specifies the shape of the array. In this case, we are creating a 1D array with 5 elements, all initialized to zero.
```

```python
arr = np.zeros((5,3))
arr
#zeros() function can also be used to create multi-dimensional arrays. In this case, we are creating a 2D array with 5 rows and 3 columns, all initialized to zero.
```

```python
arr = np.ones(5)
arr
#ones() function is used to create an array filled with ones. It takes a single argument, which specifies the shape of the array. In this case, we are creating a 1D array with 5 elements, all initialized to one.
```

```python
arr = np.ones((5,3))
arr
#ones() function can also be used to create multi-dimensional arrays. In this case, we are creating a 2D array with 5 rows and 3 columns, all initialized to one.
```

```python
arr = np.linspace(0,1,5)
arr
```

```python
arr = np.linspace(0,6,5, endpoint=False, dtype=int)
arr
#The `linspace()` function is used to create an array of evenly spaced values over a specified range. It takes several arguments, including the start and stop values, the number of samples to generate, and optional parameters such as `endpoint` and `dtype`. In this case, we are creating an array of 5 evenly spaced integers between 0 and 6 (exclusive), with the `endpoint` parameter set to `False`, which means that the stop value (6) will not be included in the output array. The `dtype` parameter is set to `int`, which specifies that the output array should contain integer values.
```

```python
array, step_size = np.linspace(0, 10, 5, retstep=True)
print(array)
print(step_size)
```

```python
rng = np.random.default_rng()
arr = rng.integers(1, 10, size=(3,4))
arr 
```

```python
rng = np.random.default_rng()
arr = rng.random((3,4))
arr 
```

```python
rng = np.random.default_rng()
arr = rng.random(5)
arr 
#rng is an instance of the default random number generator provided by NumPy. It is used to generate random numbers in a more flexible and efficient way compared to the older `np.random` module. The `rng.random()` method generates an array of random numbers drawn from a uniform distribution over the interval [0, 1). In this case, we are generating a 1D array with 5 random numbers.
```

```python
arr = rng.uniform(5, 15, size=(3, 4))
arr
#uniform() function is used to create an array of random numbers drawn from a uniform distribution. It takes three arguments: low, high, and size. The low argument specifies the lower bound of the distribution, the high argument specifies the upper bound of the distribution, and the size argument specifies the shape of the output array. In this case, we are creating a 2D array with 3 rows and 4 columns, with random numbers drawn from a uniform distribution between 5 and 15.
```

```python
arr = rng.normal(0, 1, size=(3, 4))
arr
#normal() function is used to create an array of random numbers drawn from a normal (Gaussian) distribution. It takes three arguments: loc, scale, and size. The loc argument specifies the mean of the distribution, the scale argument specifies the standard deviation of the distribution, and the size argument specifies the shape of the output array. In this case, we are creating a 2D array with 3 rows and 4 columns, with random numbers drawn from a normal distribution with a mean of 0 and a standard deviation of 1.
```

```python
a.min()
#min() function is used to find the minimum value in an array. It returns the smallest element in the array.
```

```python
int(a.min())
#int() function is used to convert the minimum value of the array to an integer data type. This is useful when we want to ensure that the minimum value is represented as an integer, especially if the array contains floating-point numbers. In this case, since the minimum value of the array `a` is already an integer (1), the conversion does not change its value.
```

```python
a.min().item()
#item() function is used to convert the minimum value of the array to a standard Python scalar type. This is useful when we want to work with the minimum value in a more general context, such as when passing it to a function that expects a scalar input. In this case, since the minimum value of the array `a` is already an integer (1), the conversion does not change its value.
```
