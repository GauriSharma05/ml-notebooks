```python
import numpy as np
```

Numpy indexing and slicing of vectors

```python
a = np.arange(11,21)
a
```

```python
a[6]
#this will give the 6th index value of the array a
```

```python
a[1:5]
#this will give the values from index 1 to 4 of the array a
```

```python
a[1:5:1]
#this will give the values from index 1 to 4 of the array a with a step size of 1 it means it will give all the values from index 1 to 4
```

```python
a[1:7:2]
#this will give the values from index 1 to 6 of the array a with a step size of 2 it means it will give the values from index 1,3,5
```

```python
a[1::2]
```

Numpy indexing and slicing of matrix

```python

a = np.arange(1,31).reshape(6,5)
a
```

```python
a[1]
#this slice out a specific row from the 2D array a.Syntax is a[row_index].
```

```python
a[1,2]
#this will give you a specific element. Syntax is a[row_index, column_index]. 
```

```python
 a[0:2,1:3]
#this will give you a specific slice of the 2D array a. Syntax is a[row_start:row_end, column_start:column_end]. and in end part for both row and column the end index is exclusive. 
```

```python
a[3:,3:]
```

```python
a[:,2]
#this will give you a specific column from the 2D array a. Syntax is a[:,column_index]. 
```
